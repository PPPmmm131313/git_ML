```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def deriv_sigmoid(x):
    fx = sigmoid(x)
    return fx * (1 - fx)
```

### три нейрона на первом скрытом слое и два - на втором

```python
class ANN_3_2:
    def __init__(self):
        # Веса
        self.w1 = np.random.normal()
        self.w2 = np.random.normal()
        self.w3 = np.random.normal()
        self.w4 = np.random.normal()
        self.w5 = np.random.normal()
        self.w6 = np.random.normal()
        self.w7 = np.random.normal()
        self.w8 = np.random.normal()
        self.w9 = np.random.normal()
        self.w10 = np.random.normal()
        self.w11 = np.random.normal()
        self.w12 = np.random.normal()
        self.w13 = np.random.normal()
        self.w14 = np.random.normal()
        self.w15 = np.random.normal()
        self.w16 = np.random.normal()
        self.w17 = np.random.normal()
        
        # Смещение
        self.b1 = np.random.normal()
        self.b2 = np.random.normal()
        self.b3 = np.random.normal()
        self.b4 = np.random.normal()
        self.b5 = np.random.normal()
        self.b6 = np.random.normal()

    def feedforward(self, x):
        h1 = sigmoid(self.w1 * x[0] + self.w2 * x[1] + self.w3 * x[2] + self.b1)
        h2 = sigmoid(self.w4 * x[0] + self.w4 * x[1] + self.w6 * x[2] + self.b2)
        h3 = sigmoid(self.w7 * x[0] + self.w8 * x[1] + self.w9 * x[2] + self.b3)

        h4 = sigmoid(self.w10 * h1 + self.w11 * h2 + self.w12 * h3 + self.b4)
        h5 = sigmoid(self.w13 * h1 + self.w14 * h2 + self.w15 * h3 + self.b5)
        
        o1 = sigmoid(self.w16 * h4 + self.w17 * h5 + self.b6)
        return o1
    

    def train(self, X, y, epochs, learning_rate):
        for epoch in range(epochs):
            for x, y_true in zip(X, y):
                # --- Прямое распространение
                sum_h1 = self.w1 * x[0] + self.w2 * x[1] + self.w3 * x[2] + self.b1
                h1 = sigmoid(sum_h1)

                sum_h2 = self.w4 * x[0] + self.w5 * x[1] + self.w6 * x[2] + self.b2
                h2 = sigmoid(sum_h2)
                
                sum_h3 = self.w7 * x[0] + self.w8 * x[1] + self.w9 * x[2] + self.b3
                h3 = sigmoid(sum_h3)
                
                sum_h4 = self.w10 * h1 + self.w11 * h2 + self.w12 * h3 + self.b4
                h4 = sigmoid(sum_h4)
                
                sum_h5 = self.w13 * h1 + self.w14 * h2 + self.w15 * h3 + self.b5
                h5 = sigmoid(sum_h5)
                
                sum_o1 = self.w16 * h4 + self.w17 * h5 + self.b6
                o1 = sigmoid(sum_o1)
                y_pred = o1

                
                # --- Расчёт частных производных
                # --- p_L_p_w1 - "частная производная L по w1"
                p_L_p_ypred = -2 * (y_true - y_pred)

                # Нейрон o1
                p_ypred_p_w16 = h4 * deriv_sigmoid(sum_o1)
                p_ypred_p_w17 = h5 * deriv_sigmoid(sum_o1)
                p_ypred_p_b6 = deriv_sigmoid(sum_o1)

                p_ypred_p_h4 = self.w16 * deriv_sigmoid(sum_o1)
                p_ypred_p_h5 = self.w17 * deriv_sigmoid(sum_o1)

                # Нейрон h4
                p_h4_p_w10 = h1 * deriv_sigmoid(sum_h4)
                p_h4_p_w11 = h2 * deriv_sigmoid(sum_h4)
                p_h4_p_w12 = h3 * deriv_sigmoid(sum_h4)
                p_h4_p_b4 = deriv_sigmoid(sum_h4)
                
                p_h4_p_h1 = self.w10 * deriv_sigmoid(sum_h4)
                p_h4_p_h2 = self.w11 * deriv_sigmoid(sum_h4)
                p_h4_p_h3 = self.w12 * deriv_sigmoid(sum_h4)

                # Нейрон h5
                p_h5_p_w13 = h1 * deriv_sigmoid(sum_h5)
                p_h5_p_w14 = h2 * deriv_sigmoid(sum_h5)
                p_h5_p_w15 = h3 * deriv_sigmoid(sum_h5)
                p_h5_p_b5 = deriv_sigmoid(sum_h5)
                
                p_h5_p_h1 = self.w13 * deriv_sigmoid(sum_h5)
                p_h5_p_h2 = self.w14 * deriv_sigmoid(sum_h5)
                p_h5_p_h3 = self.w15 * deriv_sigmoid(sum_h5)
                
                # Нейрон h1
                p_h1_p_w1 = x[0] * deriv_sigmoid(sum_h1)
                p_h1_p_w2 = x[1] * deriv_sigmoid(sum_h1)
                p_h1_p_w3 = x[2] * deriv_sigmoid(sum_h1)
                p_h1_p_b1 = deriv_sigmoid(sum_h1)
                
                # Нейрон h2
                p_h2_p_w4 = x[0] * deriv_sigmoid(sum_h2)
                p_h2_p_w5 = x[1] * deriv_sigmoid(sum_h2)
                p_h2_p_w6 = x[2] * deriv_sigmoid(sum_h2)
                p_h2_p_b2 = deriv_sigmoid(sum_h2)
                
                # Нейрон h3
                p_h3_p_w7 = x[0] * deriv_sigmoid(sum_h3)
                p_h3_p_w8 = x[1] * deriv_sigmoid(sum_h3)
                p_h3_p_w9 = x[2] * deriv_sigmoid(sum_h3)
                p_h3_p_b3 = deriv_sigmoid(sum_h3)
                
                
                # --- Обновление весов и смещения ---
                # Нейрон h1
                self.w1 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h1 * p_h1_p_w1 + p_ypred_p_h5 * p_h5_p_h1 * p_h1_p_w1 ) 
                self.w2 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h1 * p_h1_p_w2 + p_ypred_p_h5 * p_h5_p_h1 * p_h1_p_w2 ) 
                self.w3 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h1 * p_h1_p_w3 + p_ypred_p_h5 * p_h5_p_h1 * p_h1_p_w3 ) 
                self.b1 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h1 * p_h1_p_b1 + p_ypred_p_h5 * p_h5_p_h1 * p_h1_p_b1 ) 

                # Нейрон h2
                self.w4 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h2 * p_h2_p_w4 + p_ypred_p_h5 * p_h5_p_h2 * p_h2_p_w4 )
                self.w5 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h2 * p_h2_p_w5 + p_ypred_p_h5 * p_h5_p_h2 * p_h2_p_w5 )
                self.w6 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h2 * p_h2_p_w6 + p_ypred_p_h5 * p_h5_p_h2 * p_h2_p_w6 )
                self.b2 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h2 * p_h2_p_b2 + p_ypred_p_h5 * p_h5_p_h2 * p_h2_p_b2 )
                
                # Нейрон h3
                self.w7 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h3 * p_h3_p_w7 + p_ypred_p_h5 * p_h5_p_h3 * p_h3_p_w7 )
                self.w8 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h3 * p_h3_p_w8 + p_ypred_p_h5 * p_h5_p_h3 * p_h3_p_w8 )
                self.w9 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h3 * p_h3_p_w9 + p_ypred_p_h5 * p_h5_p_h3 * p_h3_p_w9 )
                self.b3 -= learning_rate * p_L_p_ypred * ( p_ypred_p_h4 * p_h4_p_h3 * p_h3_p_b3 + p_ypred_p_h5 * p_h5_p_h3 * p_h3_p_b3 )

                # Нейрон h4
                self.w10 -= learning_rate * p_L_p_ypred * p_ypred_p_h4 * p_h4_p_w10
                self.w11 -= learning_rate * p_L_p_ypred * p_ypred_p_h4 * p_h4_p_w11
                self.w12 -= learning_rate * p_L_p_ypred * p_ypred_p_h4 * p_h4_p_w12
                self.b4  -= learning_rate * p_L_p_ypred * p_ypred_p_h4 * p_h4_p_b4
                
                # Нейрон h5
                self.w13 -= learning_rate * p_L_p_ypred * p_ypred_p_h5 * p_h5_p_w13
                self.w14 -= learning_rate * p_L_p_ypred * p_ypred_p_h5 * p_h5_p_w14
                self.w15 -= learning_rate * p_L_p_ypred * p_ypred_p_h5 * p_h5_p_w15
                self.b5  -= learning_rate * p_L_p_ypred * p_ypred_p_h5 * p_h5_p_b5
                
                # Нейрон o1
                self.w16 -= learning_rate * p_L_p_ypred * p_ypred_p_w16
                self.w17 -= learning_rate * p_L_p_ypred * p_ypred_p_w17 
                self.b6  -= learning_rate * p_L_p_ypred * p_ypred_p_b6

                # Расчёт ошибки и вывод на экран (каждую 10-ю эпоху)
                if epoch % 50 == 0:
                    y_preds = np.apply_along_axis(self.feedforward, 1, X)
                    loss = mse_loss(y, y_preds)
                    print("Epoch %d loss: %.3f" % (epoch, loss))
```
